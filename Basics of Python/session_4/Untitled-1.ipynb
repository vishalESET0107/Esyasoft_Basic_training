```python
x = [2,4,5,6,8]
y = [0,6,2,8,2]

fig, ax = plt.subplots() # subplots?

ax.plot(x,y,marker = 'o',label ='data points')
ax.set_title ('matplotlib ')

ax.set_xlabel ('x.axis')
ax.set_ylabel ('y.axis')

ax.legend('hello')


plt.show()
```

```python
import pandas as pd

import numpy as np


arr = np.array([10,20,30])
fig, ax = plt.subplots()
df = pd.DataFrame({"numbers" : arr}) #why this? dataframe and all .and first of all whats a dataframe
plt.bar(df.index , df["numbers"]    )

ax.set_title('wassup')
ax.set_xlabel ('x.axis')
ax.set_ylabel ('y.axis')


plt.show()
```
